```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the uploaded dataset
df = pd.read_csv("population.csv", skiprows=4)

# Display the first few rows and the column names
df.head(), df.columns
```

```python
# Extract relevant columns: Country Name and 2022 population
df_2022 = df[['Country Name', '2022']].dropna()

# Convert population values to numeric just in case
df_2022['2022'] = pd.to_numeric(df_2022['2022'])

# Get top 10 countries by population in 2022
top10_2022 = df_2022.sort_values(by='2022', ascending=False).head(10)

# Create bar chart
plt.figure(figsize=(12, 6))
bars = plt.bar(top10_2022['Country Name'], top10_2022['2022'] / 1e6, color='mediumseagreen')
plt.title('Top 10 Most Populous Countries (2022)', fontsize=16)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Population (in millions)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add labels on bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 5, f'{yval:.0f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
